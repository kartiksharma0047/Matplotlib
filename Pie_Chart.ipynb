# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Plot** (Language Pie)


```python
x=[10,20,15,55]
y=["C","C++","C#","Python"]
ex=[0.2,0.0,0.0,0.0]
color=['r','g','b','y']
plt.pie(x,labels=y,explode=ex,colors=color,autopct="%0.1f%%",shadow=True,radius=1,startangle=0,textprops={"fontsize":11},wedgeprops={'linewidth':3})
plt.title("Basic Pie")
plt.legend(loc=4)
plt.show()
```

**Second Plot** (Dot Pie Chart)

```python
plt.pie([1],radius=0.2)
plt.show()
```

**Third Plot** (Donut Pie Chart)

```python
x=[10,20,15,55]
x1=[40,30,20,10]
y=["C","C++","C#","Python"]
ex=[0.1,0.0,0.0,0.0]
color=['r','g','b','y']
plt.pie(x,labels=y,radius=1,colors=color,explode=ex)
plt.pie([1],radius=0.7,colors="white")
plt.show()
```
