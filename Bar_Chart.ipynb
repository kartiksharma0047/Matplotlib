# Bar Plot

```python
import matplotlib.pyplot as plt
```

**First Plot** (Language Popularity)

```python
x = ["Python", "C", "C++", "Java", "Javascript"]
y = [85, 60, 70, 82, 76]
colors = ["#3776AB", "#A8B9CC", "#00599C", "#5382A1", "#F7DF1E"]

# Pick unique edge colors for each bar
edge_colors = ["#264653", "#2a9d8f", "#e76f51", "#f4a261", "#e9c46a"]
plt.bar(
    x,
    y,
    width=0.5,
    color=colors,
    edgecolor=edge_colors,
    linewidth=3,
    linestyle=":",
    alpha=0.8,
)

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Language Index", fontsize=15)
plt.show()
```

**Second Plot** (Multi-Bar Graph Overlapped)

```python
x = ["Microsoft", "Xebia", "Google", "X", "OpenAI"]
y = [60, 20, 46, 78, 36]
z = [10, 15, 30, 20, 5]
plt.xlabel("Compnay")
plt.ylabel("Out of 100")
plt.title("P/L Chart",fontsize=15)
plt.bar(x,y,color="green",label="Revenue",width=0.5)
plt.bar(x,z,color="red",label="Loss",width=0.5)
plt.legend()
plt.show()
```

**Third Plot** (Multi-Bar Graph Side-By-Side)

```python
import numpy as np

x = ["Microsoft", "Xebia", "Google", "X", "OpenAI"]
y = [60, 20, 46, 78, 36]
z = [10, 15, 30, 20, 5]

width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Company")
plt.ylabel("Out of 100")
plt.title("P/L Chart", fontsize=15)

# Revenue bars at original positions
plt.bar(p, y, color="green", label="Revenue", width=width)

# Loss bars shifted right by 'width'
plt.bar(p1, z, color="red", label="Loss", width=width)

plt.xticks([j + width/2 for j in p], x,rotation=20)  # Center the labels
plt.legend(loc="upper left")
plt.show()
```

**Forth Plot** (Horizontal Bars)

```python
x = ["Microsoft", "Xebia", "Google", "X", "OpenAI"]
y = [60, 20, 46, 78, 36]
z = [10, 15, 30, 20, 5]

width = 0.3
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Out of 100")
plt.ylabel("Company")
plt.title("P/L Chart", fontsize=15)

plt.barh(p, y, color="green", label="Revenue", height=width)
plt.barh(p1, z, color="red", label="Loss", height=width)

plt.yticks([j + width/2 for j in p], x)

plt.legend()
plt.show()
```
