# Histogram Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Plot**(Basic)

```python
length = 60
x = np.random.randint(1, 13, size=length)
bins = [y for y in range(1, 13)]

plt.hist(x, color="blue", bins=bins, edgecolor="black",bottom=1)
plt.title("Student Data", fontsize=15)
plt.xlabel("Classes")
plt.ylabel("Avg Student Passed")
plt.show()
```

**Second Plot** (cumulative histogram)

```python
length = 60
x = np.random.randint(1, 13, size=length)
bins = [y for y in range(1, 13)]

plt.hist(x, color="blue", bins=bins, edgecolor="black",cumulative=-1,rwidth=0.6)
plt.title("Student Data", fontsize=15)
plt.xlabel("Classes")
plt.ylabel("Avg Student Passed")
plt.show()
```

**Third plot** (Type step)

```python
length = 60
x = np.random.randint(1, 13, size=length)
bins = [y for y in range(1, 13)]

plt.hist(x, color="blue", bins=bins, edgecolor="black",histtype="step",orientation="horizontal",label="Design")
plt.title("Student Data", fontsize=15)
plt.xlabel("Classes")
plt.ylabel("Avg Student Passed")
plt.legend()
plt.show()
```

**Fourth Plot** (log Table)

```python
length = 60
x = np.random.randint(1, 13, size=length)
bins = [y for y in range(1, 13)]

plt.hist(x, color="blue", bins=bins, edgecolor="black",bottom=1,log=True)
plt.title("Student Data", fontsize=15)
plt.xlabel("Classes")
plt.ylabel("Avg Student Passed")
plt.show()
```

**Fifth Plot** (Seperate Line)

```python
length = 60
x = np.random.randint(1, 13, size=length)
bins = [y for y in range(1, 13)]

plt.hist(x, color="blue", bins=bins, edgecolor="black",alpha=0.5)
plt.title("Student Data", fontsize=15)
plt.xlabel("Classes")
plt.ylabel("Avg Student Passed")
plt.axvline(10,color="red",label="Higher Education")
plt.legend()
plt.show()
```
