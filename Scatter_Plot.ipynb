# Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First plot** (Basic Plot)

```python
length = 30
x = np.random.randint(1, 13, size=length)
y = np.random.randint(1, 31, size=length)
colors = ['red' if val % 2 != 0 else 'blue' for val in y]
Size = [50 if val % 2 != 0 else 70 for val in y]
plt.title("Employee Leave",fontsize=15)
plt.xlabel("Months")
plt.ylabel("Leave taken Dates")
plt.scatter(x,y,color=colors,s=Size,marker="^",edgecolors="black")
plt.show()
```

**Second plot** (cmap Added)

```python
length = 30
x = np.random.randint(1, 13, size=length)
y = np.random.randint(1, 31, size=length)

o2_levels = np.random.uniform(85, 100, size=length)

Size = 40

plt.figure(figsize=(10, 6))
sc = plt.scatter(x, y, c=o2_levels, s=Size, cmap="RdYlGn_r")

cbar = plt.colorbar(sc)
cbar.set_label('O₂ Level (%)')

plt.title("Patient O₂ Levels by Month & Day", fontsize=15)
plt.xlabel("Months")
plt.ylabel("Dates")
plt.show()
```

**Third plot** (Multi-Plots in One)

```python
import numpy as np
import matplotlib.pyplot as plt

length = 30

# O₂ plot
x1 = np.random.randint(1, 13, size=length)
y1 = np.random.randint(1, 31, size=length)
o2_levels = np.random.uniform(85, 100, size=length)

# Humidity plot — completely independent random X/Y
x2 = np.random.randint(1, 13, size=length)
y2 = np.random.randint(1, 31, size=length)
humidity = np.random.uniform(40, 90, size=length)

Size = 40

plt.figure(figsize=(12, 7))

# O₂ scatter
sc1 = plt.scatter(
    x1, y1,
    c=o2_levels,
    s=Size,
    cmap="RdYlGn_r",
    edgecolor='k',
    label='O₂ Levels'
)

# Humidity scatter
sc2 = plt.scatter(
    x2, y2,
    c=humidity,
    s=Size,
    cmap="PuBu",
    marker='s',
    edgecolor='k',
    label='Humidity'
)

# Colorbars
cbar1 = plt.colorbar(sc1, ax=plt.gca(), fraction=0.046, pad=0.04)
cbar1.set_label('O₂ Level (%)')

cbar2 = plt.colorbar(sc2, ax=plt.gca(), fraction=0.046, pad=0.08)
cbar2.set_label('Humidity (%)')

plt.title("Patient O₂ Levels & Humidity by Month & Day", fontsize=15)
plt.xlabel("Months")
plt.ylabel("Dates")
plt.legend(['O₂ Levels', 'Humidity'], loc='upper right')
plt.grid(True)
plt.show()
```
