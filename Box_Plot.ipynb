# Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np
# Box plot have 2 things in it - (1) Box Part (Median of 3 points make this) (2) Whisker Part (rest remaining line)
```

**First Plot** (Basic Single plot)

```python
limit=10
x=np.random.randint(1,70,size=limit)
x=np.append(x,100)  #outler will be visisble if a big value is enetered
plt.boxplot(x,notch=False,vert=True,widths=0.08,tick_labels=["python"],patch_artist=False,
            showmeans=True,sym="g+",boxprops=dict(color="red"),capprops=dict(color="blue"),whiskerprops=dict(color="yellow"),flierprops=dict(markeredgecolor="g"))
plt.show()
```

**Second Plot** (Multiple Box Plot)

```python
limit=10
x=np.random.randint(1,70,size=limit)
x1=np.random.randint(1,70,size=limit)
x2=np.random.randint(1,70,size=limit)
x=np.append(x,100)  #outler will be visisble if a big value is enetered
y=[x,x1,x2]
plt.boxplot(y,notch=False,vert=True,widths=0.08,tick_labels=["python","C++","Java"],patch_artist=True,
            showmeans=True,sym="g+",boxprops=dict(color="red"),capprops=dict(color="purple"),whiskerprops=dict(color="black"),flierprops=dict(markeredgecolor="g"))
plt.show()
```
