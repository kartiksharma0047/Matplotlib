# Fill Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Fill** (Fill Line Plot)

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="red",label="Python")
plt.legend(loc=2)
plt.fill_between(x=[2,4],y1=2,y2=4,color="pink")
plt.show()
```

**Second Fill** (Basic)

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="red",label="Python")
plt.legend(loc=2)
plt.fill_between(x,y,color="grey",where=(x>=2)&(x<=4),alpha=0.4)
plt.show()
```
