# Stack Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Plot** (Basic)

```python
x=[1,2,3,4,5]
area1=np.random.randint(1,15,size=5)
area2=np.random.randint(1,15,size=5)
area3=np.random.randint(1,15,size=5)
labels=["Area 1","Area 2","Area 3"]
plt.stackplot(x,area1,area2,area3,labels=labels,colors=['red','green','blue'],baseline="zero")
plt.legend(loc=2)
plt.show()
```

**Second Plot** (Change baseline)

```python
x=[1,2,3,4,5]
area1=np.random.randint(1,15,size=5)
area2=np.random.randint(1,15,size=5)
area3=np.random.randint(1,15,size=5)
labels=["Area 1","Area 2","Area 3"]
plt.stackplot(x,area1,area2,area3,labels=labels,colors=['red','green','blue'],baseline="sym")
plt.legend(loc=2)
plt.show()
```
