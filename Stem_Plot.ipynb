# Stem Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Plot** (Basic Plot)

```python
limit=10
x=np.random.randint(-10,20,size=limit)
y=np.random.randint(-10,20,size=limit)
plt.stem(x,y,linefmt="b:",markerfmt="r*",basefmt="black",label="python")
plt.legend()
plt.show()
```
