# Step Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

**First Plot** (Basic Plot)

```python
x=[1,2,3,4,5,6]
y=[1,5,3,2,5,3]
plt.step(x,y,marker="^",ms=10,mfc="red",label="Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Python")
plt.grid()
plt.legend(loc=1)
plt.show()
```
